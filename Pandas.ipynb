# **Pandas Exercises**

1. Cargar el archivo CSV

```python
import pandas as pd
students = pd.read_csv("students.csv")
```

2. Primeras 5 filas

```python
print(students.head())
```

3. Mostrar estructura del DataFrame

```python
print(students.info())
```

4. Mostrar la columna *name*

```python
print(students["name"])
```

5. Filtrar estudiantes mayores de 18

```python
print(students[students["age"] > 18])
```

6. Filtrar score > 80 y grade == "A"

```python
print(students[(students["score"] > 80) & (students["grade"] == "A")])
```

7. Crear columna *passed*

```python
students["passed"] = students["score"] >= 60
```

8. Promedio de score

```python
print(students["score"].mean())
```

9. Promedio por género

```python
print(students.groupby("gender")["score"].mean())
```

10. Contar estudiantes por ciudad

```python
print(students["city"].value_counts())
```

11. Ordenar por score descendente

```python
print(students.sort_values(by="score", ascending=False))
```

12. Estudiante(s) con mayor score

```python
max_score = students["score"].max()
print(students[students["score"] == max_score])
```

13. Reemplazar valores faltantes con la media

```python
students["score"] = students["score"].fillna(students["score"].mean())

print(students)
```

14. Nuevo DataFrame con columnas específicas

```python
new_df = students[["name", "score", "city"]]

print(new_df)
```

15. Guardar en CSV

```python
new_df.to_csv("output.csv", index=False)

print("Archivo guardado correctamente")
```

16. BONUS: Tabla dinámica

```python
pivot_table = pd.pivot_table(
    students,
    values="score",
    index="city",
    columns="gender",
    aggfunc="mean"
)

print(pivot_table)
```
